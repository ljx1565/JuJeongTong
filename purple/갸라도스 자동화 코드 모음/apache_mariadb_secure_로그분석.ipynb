```python
#Apache Error Log Parser
# /var/log/httpd/error.log 에 저장되는 로그 분석임
import re

def parse_apache_error(log_line: str):
    pattern = r'\[(?P<time>.*?)\] \[(?P<module>.*?):(?P<level>\w+)\] (?:\((?P<code>.*?)\) )?(?P<msg>.*)'
    m = re.match(pattern, log_line)

    if not m:
        return None

    return m.groupdict()


def print_apache_error(log_data: str):
    # 💡 텍스트 덩어리를 줄바꿈(\n) 기준으로 쪼갭니다.
    lines = log_data.strip().split('\n')
    
    # 💡 쪼개진 각 줄을 for문으로 하나씩 반복 처리합니다.
    for log_line in lines:
        if not log_line.strip(): # 빈 줄은 건너뛰기
            continue
            
        data = parse_apache_error(log_line)
        
        if not data:
            print(f"Parse Error: {log_line}")
            print("-" * 50)
            continue

        print(f"[WEB ERROR]")
        print(f"Time   : {data['time']}")
        print(f"Module : {data['module']}")
        print(f"Level  : {data['level']}")
        print(f"Code   : {data.get('code')}")
        print(f"Message: {data['msg']}")
        print("-" * 50) # 💡 보기 좋게 구분선 추가

apahce_error_log = """[Sun May 31 00:00:46.407132 2026] [lbmethod_heartbeat:notice] [pid 427062:tid 427062] AH02282: No slotmem from mod_heartmonitor
[Sun May 31 00:00:46.408098 2026] [mpm_event:notice] [pid 427062:tid 427062] AH00489: Apache/2.4.62 (Rocky Linux) configured -- resuming normal operations
[Sun May 31 00:00:46.408117 2026] [core:notice] [pid 427062:tid 427062] AH00094: Command line: '/usr/sbin/httpd -D FOREGROUND'"""
print_apache_error(apahce_error_log)
```

```python
#PHP-FPM Log Parser
# /var/log/php-fpm/error.log 에 저장되는 로그 분석임
import re

def parse_php_fpm(log_line: str):
    pattern = r'\[(?P<time>.*?)\] (?P<level>\w+): (?P<msg>.*)'
    m = re.match(pattern, log_line)

    if not m:
        return None

    return m.groupdict()


def print_php_fpm(log_data: str):
    # 💡 여러 줄의 텍스트를 줄바꿈(\n) 기준으로 쪼갭니다.
    lines = log_data.strip().split('\n')
    
    # 💡 쪼개진 로그를 한 줄씩 꺼내서 반복합니다.
    for log_line in lines:
        if not log_line.strip(): # 빈 줄은 무시
            continue
            
        data = parse_php_fpm(log_line)
        if not data:
            print(f"Parse Error: {log_line}")
            print("-" * 40)
            continue

        print(f"[PHP-FPM]")
        print(f"Time : {data['time']}")
        print(f"Level: {data['level']}")
        print(f"Msg  : {data['msg']}")
        print("-" * 40) # 💡 로그 간 구분을 위한 줄 긋기

php_fpm_log = """[31-May-2026 00:00:26] NOTICE: error log file re-opened"""
print_php_fpm(php_fpm_log)
```

```python
#MariaDB Audit Log Parser
# 테스트를 위해 MariaDB의 Audit log 저장 플러그인을 활성화함
# /var/lib/mysql/server_audit.log 에 저장되는 로그 분석임
def parse_mariadb_audit(log_line: str):
    """단일 Audit 로그 라인(CSV 형식)을 파싱합니다."""
    # Audit 로그는 항상 쉼표(,)로 구분된 10개의 필드를 가집니다.
    fields = log_line.split(',')
    
    # 필드가 10개가 아니라면 잘못된 포맷이므로 예외 처리
    if len(fields) < 10:
        return None
        
    return {
        'time': fields[0],         # 시간 (YYYYMMDD HH:MM:SS)
        'serverhost': fields[1],   # 서버 호스트명
        'user': fields[2],         # DB 접속 유저명
        'clienthost': fields[3],   # 접속을 시도한 클라이언트 IP/호스트
        'conn_id': fields[4],      # 연결 ID
        'query_id': fields[5],     # 쿼리 ID
        'operation': fields[6],    # 수행한 작업 (CONNECT, DISCONNECT, QUERY 등)
        'database': fields[7],     # 선택된 데이터베이스 명
        'object': fields[8],       # 건드린 테이블 명
        'retcode': fields[9]       # 결과 코드 (0이면 성공, 1045 등은 에러)
    }


def print_mariadb_audit(log_data: str):
    """여러 줄의 Audit 로그 데이터를 보기 좋게 출력합니다."""
    lines = log_data.strip().split('\n')
    
    for line in lines:
        if not line.strip():
            continue
            
        data = parse_mariadb_audit(line)
        if not data:
            print(f"[파싱 실패] {line}")
            continue

        # 작업(Operation) 종류에 따라 아이콘을 다르게 표시 (시각적 관제 용도)
        op = data['operation']
        if op == 'FAILED_CONNECT':
            icon = "🚨" # 해킹/실패 알림
        elif op in ['CONNECT', 'DISCONNECT']:
            icon = "🔌" # 단순 연결
        else:
            icon = "💾" # 쿼리/데이터 접근
            
        print(f"[{icon} MariaDB AUDIT: {op}]")
        print(f"Time    : {data['time']}")
        print(f"User    : {data['user']} (IP: {data['clienthost']})")
        
        # 빈 값이 아닐 때만 타겟 DB/테이블 출력
        if data['database'] or data['object']:
            print(f"Target  : DB({data['database']}) / Table({data['object']})")
            
        # 결과 코드 분석
        status_msg = "🟢 성공" if data['retcode'].strip() == '0' else f"🔴 실패 (Code: {data['retcode']})"
        print(f"Status  : {status_msg}")
        print("-" * 45)

db_log = """20260601 14:50:55,localhost.localdomain,root,localhost,1845,0,DISCONNECT,,,0"""
print_mariadb_audit(db_log)
```

```python
#Linux Auth Log Parser (SSH / login / vsftpd)
# /var/log/secure 에 저장되는 로그 분석임
import re

def parse_auth_log(log_line: str):
    """단일 로그 라인을 파싱하여 딕셔너리로 반환합니다."""
    pattern = r'(?P<month>\w{3})\s+(?P<day>\d+)\s+(?P<time>\d{2}:\d{2}:\d{2}) (?P<host>\S+) (?P<service>\w+)\[(?P<pid>\d+)\]: (?P<msg>.*)'
    m = re.match(pattern, log_line)

    if not m:
        return None

    data = m.groupdict()

    ip_match = re.search(r'from (?P<ip>\d+\.\d+\.\d+\.\d+)', data['msg'])
    user_match = re.search(r'for (?P<user>\w+)', data['msg'])

    data['ip'] = ip_match.group('ip') if ip_match else None
    data['user'] = user_match.group('user') if user_match else None

    return data


def print_auth_log(log_data: str):
    """
    여러 줄의 로그 텍스트를 통째로 받아서 알아서 파싱 후 출력합니다.
    """
    # 텍스트 덩어리를 줄바꿈(\n) 기준으로 쪼갭니다.
    lines = log_data.strip().split('\n')
    
    for line in lines:
        if not line.strip(): # 빈 줄은 건너뛰기
            continue
            
        data = parse_auth_log(line)
        if not data:
            print(f"Parse Error (패턴 불일치): {line}")
            print("-" * 40)
            continue

        print(f"[AUTH LOG]")
        print(f"Time   : {data['month']} {data['day']} {data['time']}")
        print(f"Host   : {data['host']}")
        print(f"Service: {data['service']}")
        print(f"PID    : {data['pid']}")
        print(f"User   : {data.get('user')}")
        print(f"IP     : {data.get('ip')}")
        print(f"Msg    : {data['msg']}")
        print("-" * 40)

acc_log = """Jun  1 10:27:38 localhost unix_chkpwd[461101]: password check failed for user (root)
Jun  1 10:27:38 localhost sshd[461099]: pam_unix(sshd:auth): authentication failure; logname= uid=0 euid=0 tty=ssh ruser= rhost=172.16.1.1  user=root
Jun  1 10:27:40 localhost sshd[461099]: Failed password for root from 172.16.1.1 port 58983 ssh2
Jun  1 10:27:47 localhost sshd[461099]: Accepted password for root from 172.16.1.1 port 58983 ssh2
Jun  1 10:27:47 localhost systemd[461108]: pam_unix(systemd-user:session): session opened for user root(uid=0) by root(uid=0)
Jun  1 10:27:47 localhost sshd[461099]: pam_unix(sshd:session): session opened for user root(uid=0) by root(uid=0)
Jun  1 10:27:47 localhost sshd[461105]: Accepted password for root from 172.16.1.1 port 58984 ssh2
Jun  1 10:27:47 localhost sshd[461105]: pam_unix(sshd:session): session opened for user root(uid=0) by root(uid=0)
Jun  1 10:29:09 localhost sshd[461099]: pam_unix(sshd:session): session closed for user root
Jun  1 10:29:09 localhost sshd[461105]: pam_unix(sshd:session): session closed for user root
Jun  1 11:06:03 localhost sshd[461836]: Accepted password for root from 172.16.10.1 port 58192 ssh2
Jun  1 11:06:03 localhost systemd[461843]: pam_unix(systemd-user:session): session opened for user root(uid=0) by root(uid=0)
Jun  1 11:06:03 localhost sshd[461836]: pam_unix(sshd:session): session opened for user root(uid=0) by root(uid=0)
Jun  1 11:06:03 localhost sshd[461841]: Accepted password for root from 172.16.10.1 port 58193 ssh2
Jun  1 11:06:03 localhost sshd[461841]: pam_unix(sshd:session): session opened for user root(uid=0) by root(uid=0)
Jun  1 11:39:19 localhost unix_chkpwd[474125]: password check failed for user (root)
Jun  1 11:39:19 localhost sshd[474099]: pam_unix(sshd:auth): authentication failure; logname= uid=0 euid=0 tty=ssh ruser= rhost=172.16.7.1  user=root
Jun  1 11:39:21 localhost sshd[474099]: Failed password for root from 172.16.7.1 port 64472 ssh2
Jun  1 11:39:25 localhost sshd[474099]: Accepted password for root from 172.16.7.1 port 64472 ssh2
Jun  1 11:39:25 localhost sshd[474099]: pam_unix(sshd:session): session opened for user root(uid=0) by root(uid=0)
Jun  1 11:39:25 localhost sshd[474158]: Accepted password for root from 172.16.7.1 port 64473 ssh2
Jun  1 11:39:25 localhost sshd[474158]: pam_unix(sshd:session): session opened for user root(uid=0) by root(uid=0)
Jun  1 11:41:13 localhost sshd[475596]: Accepted password for root from 172.16.7.1 port 64474 ssh2
Jun  1 11:41:13 localhost sshd[475596]: pam_unix(sshd:session): session opened for user root(uid=0) by root(uid=0)
Jun  1 11:41:14 localhost sshd[475596]: pam_unix(sshd:session): session closed for user root
Jun  1 12:19:27 localhost unix_chkpwd[503233]: password check failed for user (root)
Jun  1 12:19:27 localhost sshd[503207]: pam_unix(sshd:auth): authentication failure; logname= uid=0 euid=0 tty=ssh ruser= rhost=172.16.1.1  user=root
Jun  1 12:19:28 localhost sshd[503207]: Failed password for root from 172.16.1.1 port 55255 ssh2
Jun  1 12:19:32 localhost sshd[503207]: Accepted password for root from 172.16.1.1 port 55255 ssh2
Jun  1 12:19:32 localhost sshd[503207]: pam_unix(sshd:session): session opened for user root(uid=0) by root(uid=0)
Jun  1 12:19:32 localhost sshd[503297]: Accepted password for root from 172.16.1.1 port 55256 ssh2
Jun  1 12:19:32 localhost sshd[503297]: pam_unix(sshd:session): session opened for user root(uid=0) by root(uid=0)
Jun  1 13:29:53 localhost sshd[503207]: pam_unix(sshd:session): session closed for user root
Jun  1 13:29:54 localhost sshd[503297]: pam_unix(sshd:session): session closed for user root
Jun  1 13:51:41 localhost sshd[595119]: Accepted password for root from 172.16.1.1 port 56558 ssh2
Jun  1 13:51:41 localhost sshd[595119]: pam_unix(sshd:session): session opened for user root(uid=0) by root(uid=0)
Jun  1 13:51:41 localhost sshd[595147]: Accepted password for root from 172.16.1.1 port 56559 ssh2
Jun  1 13:51:41 localhost sshd[595147]: pam_unix(sshd:session): session opened for user root(uid=0) by root(uid=0)"""

print_auth_log(acc_log)
```
